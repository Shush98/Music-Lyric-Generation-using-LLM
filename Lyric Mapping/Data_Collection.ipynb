```python
import pandas as pd
import random

# Sample dataframe with lyrics
# data = {
#     'Song': ['Song 1', 'Song 2', 'Song 3'],
#     'Lyrics': ["Lyrics of song 1 line 1\nLyrics of song 1 line 2\nLyrics of song 1 line 3",
#                "Lyrics of song 2 line 1\nLyrics of song 2 line 2\nLyrics of song 2 line 3",
#                "Lyrics of song 3 line 1\nLyrics of song 3 line 2\nLyrics of song 3 line 3"]
# }

df = lyrics_df[lyrics_df['genre']=='pop']

# Function to extract random line from lyrics and replace it with <POS> tag
def extract_random_line_and_replace(lyrics):
    lines = lyrics.split('\n')
    random_index = random.randint(0, len(lines)-1)
    replaced_lines = [line if i != random_index else '<MISSING>' for i, line in enumerate(lines)]
    return '\n'.join(replaced_lines), lines[random_index]

# Apply the function to create a new column with replaced random lines
random_lines = []
for lyrics in df['Lyric']:
    replaced_lyrics, random_line = extract_random_line_and_replace(lyrics)
    df['Lyric'] = df['Lyric'].replace(lyrics, replaced_lyrics)
    random_lines.append(random_line)
df['Random_Line'] = random_lines

print(df)
```

```python
import pandas as pd

def clean_and_split_lyrics(df):
    # Clean up the text by removing "\r\n\r\n"
    df['Lyric'] = df['Lyric'].str.replace(r'\r\n\r\n', '\n')

    # Split each row into multiple rows of 6 lines each
    df['Group'] = df.index // 6

    # Extract the middle 2 lines from those 6 lines and replace them by a <MISSING> tag
    df['Missing'] = df.groupby('Group')['Lyric'].transform(lambda x: '<MISSING>').shift(-2)
    df.loc[df.index % 6 >= 2, 'Lyric'] = '<MISSING>'

    # Store the 2 extracted lines in a new column
    df['Extracted'] = df.groupby('Group')['Lyric'].transform(lambda x: x.iloc[2:4].tolist() if len(x) >= 6 else None).shift(-2)

    # Drop unnecessary columns
    df.drop(columns=['Group'], inplace=True)

    return df

# Example usage
data = {'Lyric': ["Let there be lions\r\n\r\nLet them free me from my skin\r\n\r\nLet there be water\r\n\r\nI wanna drown and touch the ground\r\n\r\nAs I drink the water of this earth\r\n\r\nAs I breathe the air in front of me\r\n\r\n\r\nLive in fear\r\n\r\nLive in fear\r\n\r\n\r\nSeek and you may find\r\n\r\nReason to hide\r\n\r\nAmong my own kind\r\n\r\nLet me be\r\n\r\nI'm falling towards the sun\r\n\r\n\r\nThis here is my solitude\r\n\r\nIt stays with me\r\n\r\nMy source of fortitude\r\n\r\n\r\nI have seen through the Hyena's eyes\r\n\r\n\r\nHigh I wanna fly\r\n\r\nTo spread my wings, fall from the sky\r\n\r\nDeep I wanna dive\r\n\r\nTo give in to the depths to feel alive\r\n\r\n"]}
df = pd.DataFrame(data)

result = clean_and_split_lyrics(df)
print(result)
```

```python
lyrics_df['Lyric'] = lyrics_df['Lyric'].str.replace('\r', '')
lyrics_df['Lyric'] = lyrics_df['Lyric'].str.replace('\n\n\n', '/N3')
lyrics_df['Lyric'] = lyrics_df['Lyric'].str.replace('\n\n', '/N2')
lyrics_df['Lyric'] = lyrics_df['Lyric'].str.replace('\n', '/N1')
```

```python
def createTriplets(song):
  if(('/N1' in song) and (song.count('/N1')>song.count('/N2'))):
    song = song.replace('/N2', '/N3')

  list_of_stanzas = song.split('/N3')

  triplets = [list_of_stanzas[i:i + 3] for i in range(len(list_of_stanzas) - 2)]
  return triplets
```

```python
def createSong(song):
  lyric = []
  missing = []
  triplets = createTriplets(song)
  for row in triplets:
    lyric.append(row[0] + '<MISSING>' + row[2])
    missing.append(row[1])
  return lyric, missing

target = lyrics_df['Lyric'].apply(createSong)
```
